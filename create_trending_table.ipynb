```python
url = "https://news.ycombinator.com/item?id=40224213"
response = requests.get(url)
soup = bs(response.text, 'html.parser')
print(soup)
```

```python
# Find all occurrences of the text
text_to_find = "Python".lower()
found_texts = soup.find_all(text=lambda text: text and text_to_find in text.lower())

# Count the number of occurrences
num_occurrences = len(found_texts)

print("Number of occurrences of '{}' is: {}".format(text_to_find, num_occurrences))
```

```python
conn = sqlite3.connect(db, timeout=30)
c = conn.cursor()

df_url = pd.read_sql_query ("SELECT urlmonth, urlyear, url FROM pageurl", conn)
df_lang = pd.read_sql_query("SELECT id, name FROM languages", conn)

for i, row in df_url.iterrows():
    url = row["url"]
    month = row["urlmonth"]
    year = row["urlyear"]
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')

    for j , lrow in df_lang.iterrows():
        lang = lrow["name"]
        lid = lrow["id"]
        text_to_find = lang.lower()
        found_texts = soup.find_all(text=lambda text: text and text_to_find in text.lower())

        # Count the number of occurrences
        num_occurrences = len(found_texts)
        if num_occurrences != 0:            
            print("Number of occurrences of '{}' in {}-20{} is: {}".format(text_to_find, month, year, num_occurrences))
        

conn.close
```
